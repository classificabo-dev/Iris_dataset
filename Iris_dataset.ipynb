```python
# üì¶ Importa√ß√£o de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.makedirs('graficos', exist_ok=True)  # cria pasta se n√£o existir


# üîç Configura√ß√µes de estilo
sns.set_style('whitegrid')
plt.style.use('ggplot')
```

```python
# üì• Leitura do dataset
df = pd.read_csv('iris_dataset.csv')

# üëÄ Visualiza√ß√£o inicial
print("üîπ Primeiras linhas:")
print(df.head())

```

```python

print("\nüîπ Informa√ß√µes do DataFrame:")
print(df.info())
```

```python
# üßπ Verifica√ß√£o e tratamento de nulos
print("\nüîπ Valores nulos por coluna:")
print(df.isnull().sum())
```

```python
# Exemplo de tratamento: remover linhas com nulos
df = df.dropna()
```

```python
# üìä Estat√≠sticas descritivas
print("\nüîπ Estat√≠sticas descritivas:")
print(df.describe())
```

```python
# üìå Distribui√ß√£o das esp√©cies (se houver coluna 'species')
if 'species' in df.columns:
    print("\nüîπ Distribui√ß√£o das esp√©cies:")
    print(df['species'].value_counts())
```

```python
# Estat√≠sticas descritivas

print(df.describe())
print(df['species'].value_counts())  # se houver coluna de esp√©cie
```

```python

# üìà Histograma das vari√°veis num√©ricas
df.hist(figsize=(10, 8), bins=15)
plt.suptitle('Distribui√ß√£o das vari√°veis num√©ricas')
plt.tight_layout()
plt.savefig('graficos/histograma_variaveis.png')  # salva o gr√°fico
plt.show()
```

```python
# üìâ Boxplot por esp√©cie (se houver coluna 'species')

if 'species' in df.columns:
    for col in df.select_dtypes(include='number').columns:
      plt.figure(figsize=(6, 4))
      sns.boxplot(data=df, x='species', y=col)
      plt.title(f'{col} por esp√©cie')
      plt.savefig(f'graficos/boxplot_{col}.png')
      plt.show()
```

```python

# üî• Mapa de calor das correla√ß√µes
correlation = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Mapa de calor das correla√ß√µes')
plt.savefig('graficos/mapa_correlacoes.png')
plt.show()
```

```python
# üîó Matriz de dispers√£o
if 'species' in df.columns:
    sns.pairplot(df, hue='species')
    plt.suptitle('Matriz de dispers√£o entre vari√°veis', y=1.02)
    plt.savefig('graficos/matriz_dispersao.png')
    plt.show()
```
